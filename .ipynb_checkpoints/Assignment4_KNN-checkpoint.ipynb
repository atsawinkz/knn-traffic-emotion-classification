```python
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
```

```python
df1 = pd.read_csv('datasets1.csv')
print("Dataset 'datasets1.csv' loaded successfully!")
df2 = pd.read_csv('datasets2.csv')
print("Dataset 'datasets2.csv' loaded successfully!")

df = pd.concat([df1, df2])
print("Datasets concatenated successfully!")

X = df.iloc[:, :-1]  # เลือกทุกคอลัมน์ยกเว้นคอลัมน์สุดท้าย
y = df.iloc[:, -1]   # เลือกคอลัมน์สุดท้ายเป็น target
print("----------Everything Successfully!----------")
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
KNN = KNeighborsClassifier()
```

```python
param_grid={
    'n_neighbors':np.arange(1,10),
    'weights':['uniform','distance'],
    'algorithm':['ball_tree', 'kd_tree', 'brute'],
    'metric':['euclidean','manhattan','minkowski','conies'],
    'p':[1,2,3]
}
Kgrid=GridSearchCV(KNN,param_grid,refit=True,verbose=3,cv=5)
Kgrid.fit(X_train,y_train)
```

```python
print('Mean Accuracy: %.2f %%'%(Kgrid.best_score_*100))
print('Best Parameters',Kgrid.best_params_)
```

```python
best_KNN = Kgrid.best_estimator_
```

```python
y_pred = best_KNN.predict(X_test)
```

```python
print(classification_report(y_test,y_pred))
print('confusion matrix:=\n',confusion_matrix(y_test,y_pred))
print('\nAccuracy:=',accuracy_score(y_test,y_pred)*100)
print('Precision:=',precision_score(y_test,y_pred,average='macro')*100)
print('Recall:=',recall_score(y_test,y_pred,average='macro')*100)
print('F1:=',f1_score(y_test,y_pred,average='macro')*100)
```
